## 2. Client Selection: Bonnie Brown (Seller)

### Profile
- **Type**: Seller
- **Situation**: Has a house, wants to move soon
- **Goal**: Achieve high profit
- **Constraint**: Property is in a middle-class neighborhood

### Why Bonnie Brown?

**1. Seller Perspective is Simpler**
- Focus on maximizing sale price
- No property search needed
- Clear success metric (profit)

**2. Middle-Class Segment is Well-Defined**
- Can be operationalized using price percentiles
- Represents the largest market segment (~50%)
- Sufficient data for reliable analysis

**3. Clear, Answerable Questions**
- When is the best time to sell? → Timing analysis
- Which areas achieve highest prices? → Location analysis
- What factors maximize profit? → Feature analysis

## 3. Key Assumptions

Since this is a fictional client, we make explicit assumptions:

### Assumption 1: Middle-Class Definition (Price-Based)

> **"Since the exact location of Bonnie Brown's property is unknown, middle-class neighborhoods are defined based on house price distributions rather than geographic or socio-economic labels."**

**Operationalization:**
- Middle-class = Houses in the 25th to 75th price percentile
- This is a data-driven, EDA-conform definition
- Based on the dataset itself, not external data

### Assumption 2: "Move Soon" = Flexible Timing
- Bonnie can wait for optimal selling season
- Timing window: within the next year
- Not an emergency sale

### Assumption 3: "High Profit" = Maximize Sale Price
- Focus on price optimization
- Consider factors that increase value
- Compare to market benchmarks
